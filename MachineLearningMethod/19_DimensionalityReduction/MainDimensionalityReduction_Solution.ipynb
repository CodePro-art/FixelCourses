```python
import numpy             as np
import pandas            as pd
import seaborn           as sns
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rc('font', **{'size' : 16})

# np.random.seed(1)
# %matplotlib notebook
```

```python
from sklearn.datasets import load_wine, load_breast_cancer
from sklearn.manifold import TSNE, SpectralEmbedding, Isomap
```

### Exercies:
* Apply Isomap, Laplacian Eigenmaps (SpectralEmbedding), and t-SNE to:
    * The breast cancer dataset
    * The wine dataset
* Plot the low-dimensional data using a scatter plot (use the true labels as color)
* Try with and without features normalization.
* You may play with the algorithms' hyper-parameters.

```python
mX, vY = load_breast_cancer(return_X_y=True)
mX     = mX - np.mean(mX, axis=0)
mX     = mX / np.std (mX, axis=0)

mX.shape, vY.shape
```

```python
lMethods = [Isomap,   SpectralEmbedding,     TSNE]
lTitle   = ['Isomap', 'Laplacian Eigenmaps', 't-SNE']

fig, _ = plt.subplots(1, 3, figsize=(16, 6))
for ii, ax in zip(range(3), fig.axes):
    oMethod = lMethods[ii]
    mZ      = oMethod().fit_transform(mX)

    ax.scatter  (mZ[:,0], mZ[:,1], c=vY, s=50, edgecolor='k')
    ax.set_title(lTitle[ii])
plt.show()
```
