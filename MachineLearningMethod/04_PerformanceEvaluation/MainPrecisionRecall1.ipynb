```python
import numpy             as np
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rc('font', **{'size':16})
# np.random.seed(1)
```

### Generate data:

```python
#-- For colab users:
# !pip install scikit-learn -U
```

```python
from sklearn.datasets import make_moons

N1        = 950
N2        = 50
mX, vY    = make_moons(n_samples=[N1, N2], noise=.1)
vY[vY==0] = -1
mX.shape, vY.shape
```

### Plot data:

```python
vIdx0 = vY == -1
plt.figure(figsize=(7, 5))
plt.scatter(mX[vIdx0,0],  mX[vIdx0,1],  s=50, c='r', edgecolor='k')
plt.scatter(mX[~vIdx0,0], mX[~vIdx0,1], s=50, c='b', edgecolor='k')

plt.show()
```

### Train SVM classifier:

```python
from sklearn.svm import SVC

oSVM  = SVC(kernel='linear').fit(mX, vY)
vHatY = oSVM.predict(mX)
```

### Plot decision boundary:

```python
#-- make grid for plot:
v0       = np.linspace(mX[:,0].min(), mX[:,0].max(), 150)
v1       = np.linspace(mX[:,1].min(), mX[:,1].max(), 150)
XX0, XX1 = np.meshgrid(v0, v1)
XX       = np.c_[XX0.ravel(), XX1.ravel()]

Z = oSVM.predict(XX)
Z = Z.reshape(XX0.shape)

plt.figure(figsize=(7, 5))
plt.contourf(XX0, XX1, Z, colors=['red', 'blue'], alpha=0.3)
plt.scatter(mX[vIdx0,0],  mX[vIdx0,1],  s=50, c='r', edgecolor='k')
plt.scatter(mX[~vIdx0,0], mX[~vIdx0,1], s=50, c='b', edgecolor='k')
plt.show()
```

### Compute precision, recall, and f1 score

```python
from sklearn.metrics import precision_recall_fscore_support

precision, recall, f1, _ = precision_recall_fscore_support(vY, vHatY, pos_label=1, average='binary')

print(f'Precision = {precision}')
print(f'Recall    = {recall}')
print(f'f1        = {f1}')
```

### Plot ROC and AUC:

```python
from sklearn.metrics import roc_curve, auc

vScore         = oSVM.decision_function(mX)
vFP, vTP, vThr = roc_curve(vY, vScore, pos_label=1)
AUC            = auc(vFP, vTP)
```

```python
plt.figure(figsize=(14, 6))

plt.subplot(1,2,1); plt.grid(True)
plt.plot(vFP,    vTP,    color='b', lw=2, label=f'ROC curve, AUC = {AUC:.3f}')
plt.plot([0, 1], [0, 1], color='k', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC')
plt.legend()

plt.subplot(1,2,2);
plt.contourf(XX0, XX1, Z, colors=['red', 'blue'], alpha=0.3)
plt.scatter(mX[vIdx0,0],  mX[vIdx0,1],  s=50, c='r', edgecolor='k')
plt.scatter(mX[~vIdx0,0], mX[~vIdx0,1], s=50, c='b', edgecolor='k')
# plt.axis('equal')

plt.show()
```

```python
from ipywidgets      import interact, IntSlider, Layout

v      = np.linspace(0, 1, 101, endpoint=True)
vThr2  = np.interp(v, vFP,  vThr)
vScore = oSVM.decision_function(XX)
mScore = vScore.reshape(XX0.shape)

def PlotRoc(thr):
    fig = plt.figure(figsize=(14, 6))
    ax  = fig.add_subplot(1, 2, 1)
    ax.plot(vFP,  vTP,  color='b', lw=3, label=f'AUC = {AUC :.3f}')
    ax.plot([0, 1], [0, 1], color='k', lw=2, linestyle='--')
    ax.axvline(x=thr/100.0, color='g', lw=2, linestyle='--')
    ax.set_xlabel('False Positive Rate')
    ax.set_ylabel('True Positive Rate')
    ax.set_title('ROC')
    ax.axis('equal')
    ax.legend()
    ax.grid()    
    
    Z    = mScore > vThr2[thr]
    ax2  = fig.add_subplot(1, 2, 2)
    ax2.contourf(XX0, XX1, Z, colors=['red', 'blue'], alpha=0.3, levels=[0, 0.5, 1.0])
    ax2.scatter(mX[vIdx0,0],  mX[vIdx0,1],  s=50, c='r', edgecolor='k')
    ax2.scatter(mX[~vIdx0,0], mX[~vIdx0,1], s=50, c='b', edgecolor='k')
#     ax2.axis('equal')
    
thrSlider = IntSlider(min=0, max=100, step=1, value=0, layout=Layout(width='80%'))
interact(PlotRoc, thr=thrSlider)
plt.tight_layout()
plt.show()
```
