```python
import numpy             as np
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rc('font', **{'size' : 16})
```

### Load data:
Download it from here:  
https://web.archive.org/web/20150922051706/http://isomap.stanford.edu/face_data.mat.Z

```python
import scipy.io

mX     = scipy.io.loadmat('./face_data.mat')['images'].T
N, D   = mX.shape
vShape = (64, 64)
mX.shape
```

### Plot few examples:

```python
K = 12
fig, _ = plt.subplots(2, 6, figsize=(20, 8))
for kk, ax in zip(range(K), fig.axes):
    idx = np.random.choice(N)
    mI  = np.reshape(mX[idx,:], vShape).T
    
    ax.imshow(mI, cmap='gray')
    ax.set_title(f'idx = {idx}')
plt.show()
```

![Isomap](https://github.com/FixelAlgorithmsTeam/FixelCourses/blob/master/MachineLearningMethod/13_DimensionalityReduction/Isomap.png?raw=true)
### Apply Isomap:

```python
from sklearn.manifold import Isomap

oIsomap  = Isomap(n_neighbors=6, n_components=2, metric='l2').fit(mX)
mZ       = oIsomap.transform(mX)
mZ.shape
```

### 2D plot of the obtained embedding:

```python
fig = plt.figure(figsize=(8, 6))
plt.scatter(mZ[:,0], mZ[:,1], s=50, c='lime', edgecolor='k')
plt.xlabel('$z_1$')
plt.ylabel('$z_2$')
plt.show()
```

```python
#-- Flipping the second axis:
mZ[:,1] *= -1
```

### Compute the farthest points:
(Just for plotting)

```python
import scipy as sp

lSet = list(range(1, N))
lIdx = [0]
for ii in range(N-1):
    mDi  = sp.spatial.distance.cdist(mZ[lIdx,:], mZ[lSet,:])
    vMin = np.min(mDi, axis=0)
    idx  = np.argmax(vMin)
    lIdx.append(lSet[idx])
    lSet.remove(lSet[idx])
```

### 2D plot of the obtained embedding (with images):

```python
fig = plt.figure(figsize=(12, 8))

size = 5
for ii in range(70):
    idx = lIdx[ii]
    x0  = mZ[idx,0] - size
    x1  = mZ[idx,0] + size
    y0  = mZ[idx,1] - size
    y1  = mZ[idx,1] + size
    mI  = np.reshape(mX[idx,:], vShape).T
    plt.imshow(mI, aspect='auto', cmap='gray', zorder=1, extent=(x0, x1, y0, y1))

plt.scatter(mZ[:,0], mZ[:,1], s=50, c='lime', edgecolor='k')
plt.xlabel('$z_1$')
plt.ylabel('$z_2$')
    
plt.show()
```
